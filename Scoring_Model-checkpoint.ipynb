Detection et gestion des valeurs manquantes

Nous examinons le nombre et le pourcentage de valeurs manquantes dans chaque colonne.

```python
 #Function to calculate missing values by column# Funct 
def missing_values_table(df):
        # Total missing values
        mis_val = df.isnull().sum()
        
        # Percentage of missing values
        mis_val_percent = 100 * mis_val / len(df)
        
        # Make a table with the results
        mis_val_table = pd.concat([mis_val, mis_val_percent], axis=1)
        
        # Rename the columns
        mis_val_table_ren_columns = mis_val_table.rename(
        columns = {0 : 'Missing Values', 1 : '% of Total Values'})
        
        # Sort the table by percentage of missing descending
        mis_val_table_ren_columns = mis_val_table_ren_columns[
            mis_val_table_ren_columns.iloc[:,1] != 0].sort_values(
        '% of Total Values', ascending=False).round(1)
        
        # Print some summary information
        print ("Your selected dataframe has " + str(df.shape[1]) + " columns.\n"      
            "There are " + str(mis_val_table_ren_columns.shape[0]) +
              " columns that have missing values.")
        
        # Return the dataframe with missing information
        return mis_val_table_ren_columns
```

Encodage des valeurs categoricales : Label Encoding, One Hot Encoding
    
Nous devons donc trouver un moyen d'encoder (représenter) ces variables sous forme de nombres avant de les transmettre au modèle. Il y a deux façons principales de procéder :
- Encodage des étiquettes : attribuer chaque catégorie unique dans une variable catégorielle avec un entier. Aucune nouvelle colonne n'est créée.
- Encodage à chaud : créer une nouvelle colonne pour chaque catégorie unique dans une variable catégorielle. Chaque observation reçoit un 1 dans la colonne pour sa catégorie correspondante et un 0 dans toutes les autres nouvelles colonnes.


```python
df_filled=application_train
numColumns = df_filled.select_dtypes(include=['int','float64']).columns
#from sklearn.preprocessing import Imputer
#imr = Imputer(missing_values='NaN', strategy='mean', axis=0)
from sklearn.impute import SimpleImputer 
imr = SimpleImputer(missing_values=np.nan, strategy='mean')

imr = imr.fit(df_filled[numColumns])
df_filled[numColumns] = imr.transform(df_filled[numColumns])
```

- créer des graphiques afin d’évaluer la performance du modèle :
    - ROC curve: La courbe ROC est similaire aux cartes de gain ou de portance en ce sens qu'elles fournissent un moyen de comparaison entre les modèles de classification. Le graphique ROC montre le taux de faux positifs (1-spécificité) sur l'axe X, la probabilité de cible=1 lorsque sa valeur réelle est 0, contre le taux de vrais positifs (sensibilité) sur l'axe Y, la probabilité de cible=1 lorsque sa valeur réelle est 1. Idéalement, la courbe monte rapidement vers le haut-gauche, ce qui signifie que le modèle a correctement prédit les cas. La ligne diagonale est pour un modèle aléatoire.


    - Gain and lift curve : La courbe gain ou la courbe lift est une mesure de l'efficacité d'un modèle de classification calculée comme le rapport entre les résultats obtenus avec et sans le modèle. Les graphiques de gain et de portance sont des aides visuelles pour évaluer les performances des modèles de classification. Cependant, contrairement à la matrice de confusion qui évalue les modèles sur l'ensemble de la population, le graphique de gain ou de levée évalue la performance du modèle sur une partie de la population.
